```python
import os
from PIL import Image

# Path to the folder containing your original images
frames_folder = 'frames'

# Path to the folder where resized images will be saved
output_folder = 'Transform_Image'

# Desired size for all images (e.g., 224x224 for many pre-trained models)
desired_size = (224, 224)

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# List of all jpg image files in the folder
image_files = [f for f in os.listdir(frames_folder) if f.endswith('.jpg')]

# Resize and save all images in the output folder
for image_file in image_files:
    image_path = os.path.join(frames_folder, image_file)
    output_path = os.path.join(output_folder, image_file)  # Set output path

    with Image.open(image_path) as img:
        resized_img = img.resize(desired_size)  # Resize image
        resized_img.save(output_path)  # Save the resized image to the output folder

print(f"Resized {len(image_files)} images and saved them to {output_folder}.")
```
